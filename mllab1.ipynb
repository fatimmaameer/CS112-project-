```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#creating  a simple dataset
data={
    'hours':[1,2,3,4,5],
    'marks':[20,27,30,35,37]
}
df=pd.DataFrame(data)

x=df['hours'].values
y=df['marks'].values
print('x:',x)
print('y:',y)

```

```python
plt.scatter(x,y)
plt.xlabel("Hours studied")
plt.ylabel("Marks Obtained")
plt.title("Hours vs Marks")
plt.show()
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
print("Mean of x : ",x_mean)
print("MEan of y : ",y_mean)
```

```python
numenator =np.sum((x-x_mean)*(y-y_mean))
denominator = np.sum((x-x_mean)**2)
a=numenator/denominator
print('slope (a):',a)
```

```python
b=y_mean - a * x_mean
print("Intercept(b):",b)
```

```python
print(f'Regression equation : y ={a:.2f}x + {b:.2f}')
```

```python
y_pred=a*x+b
print("Predicted Values",y_pred)
```

```python
#New array of unseen X values
z=np.array([4,5,6,7])
y_pred_z=a*z+b
#predict y for z
print('predicted values for z:',y_pred_z)
```

```python
error=y-y_pred
print("Residual : ",error)
```

```python
#mean sq error
mse=np.mean(error**2)
print("Mean squared error")
```
